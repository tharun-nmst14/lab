```python

import numpy as np
import matplotlib.pyplot as plt

# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
plt.subplot(2,1,1)
plt.legend("data1")

plt.title("Scatter plot of data")
# Plot the points using matplotlib
plt.plot(x, y,linestyle="dashed",linewidth=0.9,color="red")

z = np.arange(0, 3 * np.pi, 0.1)
w=np.cos(z)

plt.plot(z,w,linestyle="dashed",linewidth=0.9,color="blue")
plt.legend("data2")




x=["catA","catB","catc","catd","catE"]
y=[13,36,88,35,50]
plt.subplot(2,1,2)
plt.xlabel("x_label")
plt.ylabel("datavalues")
plt.title("Bar plot of different data")
plt.bar(x,y,width=0.8,color="green")


plt.show()

```
